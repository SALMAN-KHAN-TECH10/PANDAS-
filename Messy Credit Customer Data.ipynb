## Messy Credit Customer Data 

```python
import pandas as pd 
```

```python
data = pd.read_csv('credit_customers.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
""" STEP FOR CLEANING THE DATA 

1.. DELETING REDUNDENT COLUMNS 
2 .. RENAMEING THE COLUMNS 
3 .. DROPPING DUPLICATES 
4 .. REMOVING THE NaN VALUES 
5 .. CHECK FOR MORE TRANSFORMATION """
```

## DELETING REDUNDENT COLUMNS 

```python
data.drop(columns=['checking_status', 'duration', 'credit_history', 'purpose',
       'employment','installment_commitment', 'personal_status',
       'other_parties','residence_since', 'property_magnitude','other_payment_plans','num_dependents', 'savings_status'], inplace= True)
```

```python
data.columns
```

```python
data
```

## RENAMEING THE COLUMNS 

```python
data_new_column_names = []

for i in data.columns:
    data_new_column_names.append(i.capitalize())
```

```python
data.columns = data_new_column_names 
```

```python
data
```

## DROPPING DUPLICATES 

```python
data.duplicated().sum()

# THE DATASET DONT HAVE AN DUPLICITED  VALUES 
```

## REMOVING THE NaN VALUES

```python
data.isna().sum().sum()

# THE DATASET DONT HAVE AN NA VALUES 
```

## RESETTING THE DATASET INDEX FOR BETTER UNDERSTANDING 

```python
data.reset_index()
```
