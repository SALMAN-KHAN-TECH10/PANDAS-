## CLEANING OF MESSY EXAMS DATA WITH THE PANDAS 

```python
import pandas as pd
```

```python
data = pd.read_csv('EXAMS_DATASET.csv')
data.head()
```

```python
""" STEP FOR CLEANING THE DATA 

1.. DELETING REDUNDENT COLUMNS 
2 .. RENAMEING THE COLUMNS 
3 .. DROPPING DUPLICATES 
4 .. REMOVING THE NaN VALUES 
5 .. CHECK FOR MORE TRANSFORMATION """
```

```python
data.columns
```

## DELETING REDUNDENT COLUMNS 

```python
data.drop(columns=['Unnamed: 0','TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild',],inplace=True)
```

```python
data
```

## RENAMEING THE COLUMNS

```python
data.rename(columns={ 'NrSiblings' : 'NUM.OF.SIBLINGS' },inplace=True)
```

```python
data
```

## DROPPING DUPLICATES

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

## REMOVING THE NaN VALUES 

```python
data.isna().sum().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum().sum()
```
